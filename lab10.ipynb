## Python vs JavaScript — Quick Reference

| Concept | JavaScript | Python |
|---------|------------|--------|
| Variable | `const dogName = 'Judy'` | `dog_name = 'Judy'` |
| Print | `console.log(dogName)` | `print(dog_name)` |
| Template string | `` `${dogName} is ${age}` `` | `f'{dog_name} is {age}'` |
| Array / List | `['Airedale', 'Beagle']` | `['Airedale', 'Beagle']` |
| Object / Dictionary | `{ name: 'Judy', age: 4 }` | `{'name': 'Judy', 'age': 4}` |
| For loop | `for (const d of dogs) { }` | `for d in dogs:` |
| Function | `function describe(dog) { }` | `def describe(dog):` |
| Array length | `dogs.length` | `len(dogs)` |
| Structure delimiter | `{ }` curly braces | **indentation** (4 spaces) |

> **The biggest mental shift:** Python uses indentation where JS uses `{}`. Everything inside a loop, function, or `if` block must be indented 4 spaces — there are no curly braces to close.

### Variables

In JavaScript you used `const` and `let`. Python skips the keyword entirely — just assign directly:

```js
// JavaScript
const dogName = 'Judy';
let breed = 'Airedale';
let age = 6;
```

```python
# Python
dog_name = 'Judy'
breed    = 'Airedale'
age      = 6
```

**Why no `const` or `let`?** In Python, a variable is just a name bound to an object in memory — there's no separate declaration step. Python is *dynamically typed*, meaning the interpreter figures out the type from the value at runtime, so you never need to declare it upfront. There's no `const` because Python doesn't enforce immutability on simple variables at the language level (the convention is `ALL_CAPS` for values you intend to treat as constants).

Two more things to notice:
- No `const` or `let` — just the name and value
- Python convention uses `snake_case` instead of `camelCase`

### f-strings

Python's answer to JavaScript template literals — put an `f` before the opening quote:

```js
// JavaScript
`${dogName} is a ${breed}, age ${age}`
```

```python
# Python
f'{dog_name} is a {breed}, age {age}'
```

**Why the `f` prefix?** An f-string is evaluated at runtime — Python scans it for `{}` expressions, evaluates each one, converts the result to a string, and inserts it in place. The format specifier after the colon (`:,` for thousands commas, `:.1f` for one decimal place) is passed to Python's built-in `format()` function under the hood, giving you precise control over how each value is displayed.

```python
f'{dog_name} weighs {weight:.1f} lbs'   # → 'Judy weighs 52.3 lbs'
```

### Lists

```js
// JavaScript
```

**Why does this feel familiar?** A Python list is a *mutable, ordered sequence* — "mutable" means you can change it after creation by adding, removing, or reordering items, just like a JavaScript array. Negative indexing works because Python automatically adds the list length to a negative index: `breeds[-1]` is literally computed as `breeds[len(breeds) - 1]` at runtime, giving you `'Beagle'`.

## Part 3 — Dictionaries

We saw in JavaScript how we could build a county object with properties like name, state, and population. A Python dictionary is the exact same concept.

### Dictionaries

```js
// JavaScript
const dogInfo = {
  name: 'Judy',
  breed: 'Airedale',
  age: 6
};
console.log(dogInfo.name);     // 'Judy'
console.log(dogInfo['breed']); // 'Airedale'
```

```python
# Python
dog_info = {
    'name':  'Judy',
    'breed': 'Airedale',
    'age':   6
}
print(dog_info['name'])   # 'Judy'
print(dog_info['breed'])  # 'Airedale'
```

**Why must keys be quoted, and why no dot notation?** A Python dictionary is a *hash map* — each key is hashed to find its storage slot, which makes lookups O(1) regardless of how large the dictionary gets. Keys must be quoted because Python requires them to be *hashable* types (strings, numbers, tuples); unlike JavaScript, they are not automatically converted from bare identifiers. Dot notation is reserved for accessing attributes on objects (classes), not plain dict keys.

Key differences from JS objects:
- Keys **must be quoted strings** — no bare `name:` syntax
- Access is always bracket notation: `dog_info['name']`

````python
# YOUR TURN — build a dictionary for your county
# Include at least: name, state, population, county_seat, founded

county_info = {
    'name': 'Bucks',
    'state': 'PA',
    'population': 630000,
    'county_seat': 'Doylestown',
    'founded': 1682
}
print(county_info['name'])
print(county_info['county_seat'])

# Print two values using bracket notation
````

## Part 4 — Lists of Dictionaries


### Lists of Dictionaries

Combine a list and a dictionary and you get the most common data shape in Python — a list of records. Each item is a dictionary describing one thing:

```python
# Python
dogs = [
    {'name': 'Judy',    'breed': 'Airedale', 'age': 6},
    {'name': 'Rex',     'breed': 'Labrador', 'age': 7},
    {'name': 'Biscuit', 'breed': 'Beagle',   'age': 2},
]
```

Accessing values:
- One item from the list: `dogs[0]` → the whole dictionary for Judy
- A field inside that item: `dogs[0]['name']` → `'Judy'`
- Inside a loop: `dog['breed']`

**Why no interface?** Python uses *duck typing* — if an object has the key you try to access, it works; Python doesn't check structure upfront. This makes it faster to write but means a typo in a key name won't be caught until that line actually runs. TypeScript's interface gives you the compile-time safety net Python skips by design.

````python
# YOUR TURN — build a list of dictionaries for your county's towns
# Look up at least 5 towns on Google Maps — right-click any point to copy coordinates
# Each dictionary needs: name, population, lat, lon

towns = [
    {'name': 'Yardley',  'population': 2585,  'lat': 40.2305,  'lon': -74.8854},
    {'name': 'Churchville',  'population': 6084,  'lat': 40.18717151207045,  'lon': -74.99063576152511},
    {'name': 'Newtown',  'population': 2277,  'lat':  40.22837666296971,  'lon': -74.93157731592152},
    {'name': 'Warminster',  'population': 3987,  'lat':  40.22837666296971,  'lon': -74.93157731592152},
    {'name': 'Doylestown',  'population': 8424,  'lat':  40.310,  'lon': -75.130},

    # add more rows...
]

print(f'Loaded {len(towns)} towns')
print(f"First town: {towns[0]['name']}")
````

## Part 5 — Functions

We've been calling functions in JavaScript this semester. Now you will define them.

```js
// JavaScript
function describe(dog) {
    return `${dog.name} is a ${dog.breed}`;
}
```

```python
# Python
def describe(dog):
    return f'{dog["name"]} is a {dog["breed"]}'
```

**Why does `def` work this way?** In Python, `def` is an executable statement that creates a function object and binds it to a name — exactly like any other assignment. This makes functions *first-class objects*: you can store them in variables, pass them as arguments, or return them from other functions. The `key=lambda t: t['population']` in the Bonus section is a direct example of passing a small function as an argument to `sorted()`.

Rules for `def`:
- No `function` keyword — just `def`
- No curly braces — indent the body 4 spaces
- `return` works exactly the same way
- `pass` is a placeholder that means "nothing here yet" — replace it with your code

### If Statements — Quick Syntax Primer

You'll need an `if` inside your functions. Here's how it compares to JavaScript:

```js
// JavaScript
if (population > 500000) {
    console.log("large county");
} else {
    console.log("smaller county");
}
```

```python
# Python
if population > 500000:
    print("large county")
else:
    print("smaller county")
```

Three things to remember:
- **No parentheses** required around the condition (they're allowed, just not needed)
- **Colon at the end** of every `if`, `elif`, and `else` line — easy to forget
- **Indentation instead of `{ }`** — the body must be indented 4 spaces

`elif` is Python's version of `else if`:

```python
if population > 500000:
    print("large county")
elif population > 100000:
    print("medium county")
else:
    print("small county")
```

````python
# most_populous_town(towns)  →  returns the dictionary of the most populous town
#
# You already wrote this logic in Part 4 — now wrap it in a reusable function

def most_populous_town(towns):
    most_populous = towns[0]

    for town in towns:
      if town['population'] > most_populous['population']:
        most_populous = town
    return most_populous

# Test it
result = most_populous_town(towns)
print(f"Most populous: {result['name']} ({result['population']:,})")
````

````python
# geographic_center(towns)  →  returns (avg_lat, avg_lon) as a tuple
#
# Compute the average latitude and average longitude across all towns.
# This is the same coordinate thinking as Lab 3b — just in Python.

def geographic_center(towns):
    total_lat = 0
    total_lon = 0
    # Loop through towns and accumulate total_lat and total_lon
    for town in towns:
      total_lat += town['lat']
      total_lon += town['lon']
    return (total_lat / len(towns), total_lon / len(towns))


# Test it — then paste the result into Google Maps to verify!
center = geographic_center(towns)
print(f'Geographic center of your county: {center[0]:.4f}, {center[1]:.4f}')
````
