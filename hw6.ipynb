**Concept:** dictionary access, adding keys, `.items()` loop

Two of Aaron Judge's worst postseason series *(source: Baseball Reference / StatMuse)*:
- **2017 ALDS vs. Cleveland** — .050 AVG, 16 strikeouts (an ALDS record). Yankees won the series.
- **2022 ALCS vs. Houston** — 1-for-16, .063 AVG, OPS .181. Yankees swept in four games.

**a)** Print the `avg` for `'2022 ALCS vs. Houston'`.

**b)** Add the 2024 World Series vs. LA Dodgers: `avg=0.222, hr=1, games=5`.
(Yankees lost 4-1; Judge had 7 strikeouts in the first three games.)

**c)** Loop with `.items()` and print each entry:
`2022 ALCS vs. Houston → 0.063 AVG | 0 HR`

> **Hint:** Use `:.3f` inside your f-string (e.g. `{stats['avg']:.3f}`) so all averages print with three decimal places — otherwise `0.050` will display as `0.05`.

**d)** Find and print the series with the **lowest** batting average.

````python
judge_tough_series = {
    '2017 ALDS vs. Cleveland': {'avg': 0.050, 'hr': 0, 'games': 5},
    '2022 ALCS vs. Houston':   {'avg': 0.063, 'hr': 0, 'games': 4},
}

# a) Print avg for 2022 ALCS:
print(f"2022 ALCS avg: {judge_tough_series['2022 ALCS vs. Houston']['avg']:.3f}")

# b) Add 2024 World Series:
judge_tough_series['2024 World Series vs. LA Dodgers'] = {'avg': 0.222, 'hr': 1, 'games': 5}

# c) Loop and print each series:
for series, stats in judge_tough_series.items():
  print(f"{series}: {stats['avg']:.3f} AVG | {stats['hr']} HR")

# d) Find lowest avg series:
lowest_series = min(judge_tough_series, key = lambda x: judge_tough_series[x]['avg'])
lowest_value = judge_tough_series[lowest_series]['avg']
print(f"\nLowest avg series: {lowest_series} ({lowest_value:.3f} AVG)")
````

**Concept:** for loop, if / elif / else, arithmetic

- `Healthy` — 40 to 50 lbs (inclusive)
- `Overweight` — above 50 lbs

Format: `42.0 lbs — Healthy`

**b)** Calculate and print her average weight using `:.1f`.

````python
checkup_weights = [38.5, 42.0, 45.0, 51.5, 44.0, 45.0]

# a) Loop and label each weight:
for weight in checkup_weights:
  if weight < 40:
    label = "Underweight"
  elif 40 <= weight <= 50:
    label = "Healthy"
  else:
    label = "Overweight"
  print(f"{weight} lbs - {label}")

# b) Calculate and print average weight:
avg_weight = sum(checkup_weights) / len(checkup_weights)
print(f"Average weight: {avg_weight:.1f} lbs")
````

**Concept:** `.upper()`, `.lower()`, `.replace()`, `.split()`, `in`

**a)** Loop through `dining_locations` and print each name in ALL CAPS using `.upper()`.

## Question 14 — Function + Search &nbsp;(8 pts)
**Concept:** `def`, loop, `return`, case-insensitive matching

Write `find_dining(locations, keyword)` that:
- Returns a **new list** of all locations whose `name` contains the keyword (case-insensitive)
- Returns an **empty list** (not `None`) if nothing matches

**a)** Call `find_dining(dining, 'the')` and print the name of each match.

**b)** Call `find_dining(dining, 'coffee')` and print the result.

**c)** Call `find_dining(dining, 'gym')`. If the list is empty, print `'No locations found.'`

````python
dining = [
    {'name': 'Cafe Nova',               'seats': 120},
    {'name': 'Wawa (Connelly Center)',   'seats': 40},
    {'name': "St. Mary's Dining Hall",  'seats': 350},
    {'name': 'The Commons',             'seats': 500},
    {'name': 'The Refectory',           'seats': 280},
    {'name': "Rosie's Coffee Hut",      'seats': 30},
    {'name': 'Campus Corner',           'seats': 80},
]

def find_dining(locations, keyword):
    results = []
    for loc in locations:
      if keyword.lower() in loc['name'].lower():
        results.append(loc)
    return results

# a) Search 'the' and print matching names:
print("Part A- searching for 'the':")
found_the = find_dining(dining, 'the')
for match in found_the:
  print(match['name'])

# b) Search 'coffee' and print:
print("\nPart B- searching for 'coffee':")
found_coffee = find_dining(dining, 'coffee')
print(found_coffee)

# c) Search 'gym' — print 'No locations found.' if empty:
print("\nPart C- searching for 'gym':")
found_gym = find_dining(dining, 'gym')
if not found_gym:
  print("No locations found.")
else:
  print(found_gym)
````

## Question 15 — Yankees Championship Drought &nbsp;(8 pts)
**Concept:** list arithmetic, loops with `range()`, finding max, if/else

The New York Yankees have won **27 World Series titles** — the most in MLB history.
Their championship years are listed below. Their last title was in 2009.

**a)** Print the total number of championships using `len()`.

**b)** Compute the gap between each consecutive pair of championships.
Use `range(len(championships) - 1)` to loop through index pairs.
Print each gap:
```
1978 → 1996: 18 years
```

**c)** Find and print the **longest** gap between championships in Yankees history.

**d)** Compute the current drought: `current_year - championships[-1]`.
Print: `The current drought is X years (and counting).`

**e)** Is the current drought the longest in Yankees history?
Print `'Yes — this is the longest drought in Yankees history.'` or
`'No — the longest was X years (YYYY–YYYY).'`

````python
championships = [
    1923, 1927, 1928, 1932, 1936, 1937, 1938, 1939, 1941, 1943,
    1947, 1949, 1950, 1951, 1952, 1953, 1956, 1958, 1961, 1962,
    1977, 1978, 1996, 1998, 1999, 2000, 2009,
]

current_year = 2026

# a) Total championships:
print("PART A- TOTAL CHAMPIONSHIPS:")
print(f"Total championships: {len(championships)}")

# b) Gap between each consecutive pair — use range(len(championships) - 1):
print("\nPART B AND C- GAPS:")
longest_historical_gap = 0     # defining var for part c
for i in range(len(championships) - 1):
  year1 = championships[i]
  year2 = championships[i + 1]
  gap = year2 - year1
  print(f"{year1} -> {year2}: {gap} years")

# c) Longest historical gap:
  if gap > longest_historical_gap:
    longest_historical_gap = gap
    years_of_gap = f"{year1} - {year2}"
print(f"Longest historical gap: {longest_historical_gap} years ({years_of_gap})")

# d) Current drought:
print("\nPART D- CURRENT DROUGHT:")
current_drought = current_year - championships[-1]
print(f"The current drought is {current_drought} years (and counting).")

# e) Is the current drought the longest?
print("\nPART E- IS CURRENT DROUGHT THE LONGEST:")
if current_drought > longest_historical_gap:
  print("Yes- this is the longest drought in Yankees history.")
else:
  print(f"No- the longest was {longest_historical_gap} years ({years_of_gap}).")
````
